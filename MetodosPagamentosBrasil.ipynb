## Questões

### Questão 1 - Carregue o dataset "brazilian_payment_methods.csv" usando o Pandas e exiba as primeiras 5 linhas.

```python
dados = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vSDZCObTOIR30-UlT7V7EVY3vOrEyuwpRNuIGBhPlxhEXcNYyVTm7BrhvNvC2KLtPkYEx-XIdDPCeob/pub?output=csv')
dados.head()
```

### Questão 2 - Troque os nomes das colunas do dataset para português:

```python
dados = dados.rename(columns= {
    'YearMonth': 'mesReferencia',
    'quantityPix': 'numeroTransacoesPix',
    'valuePix': 'valorTotalTransacoesPix',
    'quantityTED': 'numeroTransacoesTED',
    'valueTED': 'valorTotalTransacoesTED',
    'quantityTEC': 'numeroTransacoesTEC',
    'valueTEC': 'valorTotalTransacoesTEC',
    'quantityBankCheck': 'numeroTransacoesCheque',
    'valueBankCheck': 'valorTotalTransacoesCheque',
    'quantityBrazilianBoletoPayment': 'numeroTransacoesBoleto',
    'valueBrazilianBoletoPayment': 'valorTotalTransacoesBoleto',
    'quantityDOC': 'numeroTransacoesDOC',
    'valueDOC': 'valorTotalTransacoesDOC',
})

dados.head()
```

### Questão 3 - Selecione apenas as colunas YearMonth, quantityPix, e valuePix e exiba as primeiras 10 linhas.

```python
dados[['mesReferencia', 'numeroTransacoesPix', 'valorTotalTransacoesPix']].head(10)
```

### Questão 13 - Calcule a correlação entre o valor total das transações PIX (valuePix) e o valor total das transações TED (valueTED). O que você pode inferir sobre a relação entre esses dois métodos de pagamento?

```python
correlacao_pix_ted = dados[['valorTotalTransacoesPix', 'valorTotalTransacoesTED']].corr()
display(correlacao_pix_ted)
```

```python
plt.figure(figsize=(5, 3))
sns.scatterplot(x=dados['valorTotalTransacoesPix'], y=dados['valorTotalTransacoesTED'])
plt.title('Relação entre Valor Total das Transações PIX e TED')
plt.xlabel('Valor Total das Transações PIX')
plt.ylabel('Valor Total das Transações TED')
plt.show()
```

### Questão 14 - Compare o crescimento percentual do valor total das transações PIX (valuePix) e TED (valueTED) ao longo do tempo.

```python
crescimento_percentual_pix = dados['valorTotalTransacoesPix'].pct_change() * 100
crescimento_percentual_ted = dados['valorTotalTransacoesTED'].pct_change() * 100
```

```python
comparacao_crescimento = pd.DataFrame({
    'AnoMes': dados['mesReferencia'],
    'Crescimento Percentual PIX': crescimento_percentual_pix,
    'Crescimento Percentual TED': crescimento_percentual_ted
})

display(comparacao_crescimento.head())
```

```python
plt.figure(figsize=(5, 3))
plt.plot(dados['Ano'], crescimento_percentual_pix, marker='', linestyle='-', label='Crescimento Percentual PIX')
plt.plot(dados['Ano'], crescimento_percentual_ted, marker='', linestyle='-', label='Crescimento Percentual TED')

plt.title('Crescimento Percentual do Valor Total das Transações PIX e TED ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('Crescimento Percentual (%)')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(0.5, -0.3), ncol=1)
plt.gca().invert_xaxis()
plt.show()
```

### Questão 15 - Calcule a participação percentual de cada método de pagamento (PIX, TED, TEC, Boleto, DOC) no valor total das transações para o último mês disponível no dataset.

```python
ultimo_mes = dados['mesReferencia'].max()
dados_ultimo_mes = dados[dados['mesReferencia'] == ultimo_mes]

total_pix = dados_ultimo_mes['valorTotalTransacoesPix'].sum()
total_ted = dados_ultimo_mes['valorTotalTransacoesTED'].sum()
total_tec = dados_ultimo_mes['valorTotalTransacoesTEC'].sum()
total_boleto = dados_ultimo_mes['valorTotalTransacoesBoleto'].sum()
total_doc = dados_ultimo_mes['valorTotalTransacoesDOC'].sum()

total_transacoes = total_pix + total_ted + total_tec + total_boleto + total_doc

participacao_pix = (total_pix / total_transacoes) * 100
participacao_ted = (total_ted / total_transacoes) * 100
participacao_tec = (total_tec / total_transacoes) * 100
participacao_boleto = (total_boleto / total_transacoes) * 100
participacao_doc = (total_doc / total_transacoes) * 100

print(f"Participação percentual de PIX: {participacao_pix:.2f}%")
print(f"Participação percentual de TED: {participacao_ted:.2f}%")
print(f"Participação percentual de TEC: {participacao_tec:.2f}%")
print(f"Participação percentual de Boleto: {participacao_boleto:.2f}%")
print(f"Participação percentual de DOC: {participacao_doc:.2f}%")
```

### Questão Extra 3 - Qual foi o método de pagamento mais utilizado em termos de quantidade de transações (quantityPix, quantityTED, quantityTEC, quantityBrazilianBoletoPayment, quantityDOC) no último ano disponível no dataset?

```python
ultimo_ano = dados['Ano'].max()
dados_ultimo_ano = dados[dados['Ano'] == ultimo_ano]

quantidade_transacoes = {
    'PIX': dados_ultimo_ano['numeroTransacoesPix'].sum(),
    'TED': dados_ultimo_ano['numeroTransacoesTED'].sum(),
    'TEC': dados_ultimo_ano['numeroTransacoesTEC'].sum(),
    'Boleto': dados_ultimo_ano['numeroTransacoesBoleto'].sum(),
    'DOC': dados_ultimo_ano['numeroTransacoesDOC'].sum()
}

metodo_mais_utilizado = max(quantidade_transacoes, key=quantidade_transacoes.get)
quantidade = quantidade_transacoes[metodo_mais_utilizado]

print(f"O método de pagamento mais utilizado no último ano ({ultimo_ano}) foi o {metodo_mais_utilizado} com {quantidade} transações.")
```

### Questão Extra 4 - Calcule o valor médio por transação para cada método de pagamento (PIX, TED, TEC, Boleto, DOC) no último mês disponível.

```python
ultimo_mes = dados['mesReferencia'].max()
dados_ultimo_mes = dados[dados['mesReferencia'] == ultimo_mes]

valor_medio_por_transacao = {
    'PIX': dados_ultimo_mes['valorTotalTransacoesPix'].sum() / dados_ultimo_mes['numeroTransacoesPix'].sum() if dados_ultimo_mes['numeroTransacoesPix'].sum() != 0 else 0,
    'TED': dados_ultimo_mes['valorTotalTransacoesTED'].sum() / dados_ultimo_mes['numeroTransacoesTED'].sum() if dados_ultimo_mes['numeroTransacoesTED'].sum() != 0 else 0,
    'TEC': dados_ultimo_mes['valorTotalTransacoesTEC'].sum() / dados_ultimo_mes['numeroTransacoesTEC'].sum() if dados_ultimo_mes['numeroTransacoesTEC'].sum() != 0 else 0,
    'Boleto': dados_ultimo_mes['valorTotalTransacoesBoleto'].sum() / dados_ultimo_mes['numeroTransacoesBoleto'].sum() if dados_ultimo_mes['numeroTransacoesBoleto'].sum() != 0 else 0,
    'DOC': dados_ultimo_mes['valorTotalTransacoesDOC'].sum() / dados_ultimo_mes['numeroTransacoesDOC'].sum() if dados_ultimo_mes['numeroTransacoesDOC'].sum() != 0 else 0
}

for metodo, valor_medio in valor_medio_por_transacao.items():
    print(f"Valor médio por transação para {metodo} no último mês ({ultimo_mes}): R${valor_medio:.2f}")
```

### Questão Extra 6 - Crie boxplots lado a lado para comparar a dispersão do valor total das transações PIX (valuePix) e TED (valueTED). Calcule a variância de cada um. Qual método de pagamento tem maior dispersão?

```python
variancia_pix = dados['valorTotalTransacoesPix'].var()
variancia_ted = dados['valorTotalTransacoesTED'].var()

print(f"Variância do valor total das transações PIX: {variancia_pix:.2f}")
print(f"Variância do valor total das transações TED: {variancia_ted:.2f}")

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.boxplot(dados['valorTotalTransacoesPix'], vert=False)
plt.title('Dispersão do Valor Total das Transações PIX')

plt.subplot(1, 2, 2)
plt.boxplot(dados['valorTotalTransacoesTED'], vert=False)
plt.title('Dispersão do Valor Total das Transações TED')

plt.tight_layout()
plt.show()
```

O TED tem uma maior dispersão nos valores das transações em comparação com o PIX.
A variância do TED (749.147.926.035,79) é significativamente maior do que a variância do PIX (387.015.775.524,30), indicando que os valores das transações TED têm maior variação, ou seja, há mais transações TED com valores muito altos ou muito baixos em comparação com o PIX.
